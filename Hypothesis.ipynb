```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/m_clean.csv')
df
```

```python
df['ratio'] = df['consume']/df['distance']
df.head()
```

```python
e10 = df[df['gas_type'] == 'E10']
e10
```

```python
e10['ratio'].mean()
```

```python
# Your code here:
# One-sided test:
# H0: SP98 gas consumption is lower than E10 on average (<= 0.44)
# H1: SP98 gas consumption is higher than E10 on average (> 0.44)
```

```python
SP98 = df[df['gas_type'] == 'SP98']
SP98.head()
```

```python
from scipy.stats import ttest_1samp
```

```python
stat, pval = ttest_1samp(SP98['ratio'], 0.44)

print('stat is  ', stat)
print('P-value for the one-tailed test is ', pval/2)
```

**Answer:**   
The P-value is less than 0.05, which means we reject $H_0:$  (SP98 gas consumption is lower than E10 on average <= 0.44)

```python
import scipy.stats

confidence_level = 0.95
degrees_freedom = len(SP98['ratio']) - 1  
sample_mean = np.mean(SP98['ratio'])

sample_standard_error = scipy.stats.sem(SP98['ratio'])

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)

print( 'Confidence interval is ', confidence_interval, '.' )
```

Based on our confidence interval values, the SP98 gas consumption is between 0.39 and 0.62.
